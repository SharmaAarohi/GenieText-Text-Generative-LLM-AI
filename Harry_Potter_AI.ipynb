```python
# Open and read the text file
with open('Harry Potter and the Sorcerers Stone.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

Cleaning pdf text

```python
import re

# Function to clean the text
def clean_text(text):
    # Remove page numbers (assuming they are isolated on lines)
    text = re.sub(r'\b\d+\b', '', text)

    # Remove special characters (keeping basic punctuation and alphanumeric)
    text = re.sub(r'[^A-Za-z0-9 .,\'"\n]+', ' ', text)

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Clean the text
cleaned_text = clean_text(text)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
from transformers import set_seed

# Set padding token to eos token if not already set
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Function to split text into chunks
def split_text_into_chunks(text, chunk_size=1024):
    tokens = tokenizer.encode(text)
    return [tokens[i:i+chunk_size] for i in range(0, len(tokens), chunk_size)]

# Tokenize and chunk the cleaned text
chunks = split_text_into_chunks(cleaned_text)

print(chunks[:1])
```

```python
# Print the length of tokens in each chunk
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {len(chunk)} tokens")
```

```python
# Generate text
import torch

set_seed(42)

def generate_text(input_ids):
    # Truncate input_ids if it exceeds the vocabulary size
    input_ids = input_ids[:, :tokenizer.model_max_length]
    print("Input IDs before generation:", input_ids)
    attention_mask = input_ids.ne(tokenizer.eos_token_id).long()  # Create attention mask
    print("Attention Mask:", attention_mask)
    output = model.generate(input_ids, attention_mask=attention_mask, max_new_tokens=100, pad_token_id=tokenizer.eos_token_id)
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
class QADataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.sep_token = tokenizer.sep_token if tokenizer.sep_token else ' '

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        question = item['question']
        answer = item['answer']
        context = item['context']
        inputs = self.tokenizer.encode_plus(
            question + self.sep_token + context,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        inputs['labels'] = self.tokenizer.encode(answer, truncation=True, padding='max_length', max_length=self.max_length, return_tensors='pt')
        return inputs
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Define the functions
def evaluate_model(model, tokenizer, question, context, max_length=512):
    # Using a common separator token
    sep_token = "<|endoftext|>"
    inputs = tokenizer.encode_plus(question + sep_token + context, return_tensors='pt')
    input_ids = inputs['input_ids']
    attention_mask = inputs.get('attention_mask')

    output = model.generate(input_ids, attention_mask=attention_mask, max_length=max_length, num_beams=4, early_stopping=True)
    answer = tokenizer.decode(output[0], skip_special_tokens=True)
    return answer

# Load pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Example evaluation with context
question = "What earned Harry his longest-ever punishment?"
context = "The escape of the Brazilian boa constrictor earned Harry his longest-ever punishment."
answer_with_context = evaluate_model(model, tokenizer, question, context)
print("Answer with context:", answer_with_context)
```

```python
def evaluate_model_without_context(model, tokenizer, question, max_length=512):
    inputs = tokenizer.encode_plus(question, return_tensors='pt')
    input_ids = inputs['input_ids']
    attention_mask = inputs.get('attention_mask')

    output = model.generate(input_ids, attention_mask=attention_mask, max_length=max_length, num_beams=4, early_stopping=True)
    answer = tokenizer.decode(output[0], skip_special_tokens=True)
    return answer

# Load pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Example evaluation without context
question = "What earned Harry his longest-ever punishment?"
answer = evaluate_model_without_context(model, tokenizer, question)
print("Answer without context:", answer)
```
